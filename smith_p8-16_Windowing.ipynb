In this problem, you will make sets of $N=1,024$ samples taken at intervals of $\Delta = 1\;{\rm ms}$, so that the total sampling time is $T= 1.024\;{\rm s}$. Recall that the Discrete Fourier Transform (DFT) algorithm assumes that the input waveform $y(t)$ has the periodicity $T$. In this problem, you'll explore what happens when this isn't true.
 a Create a data of 1,024 samples of the continuous wave $y= \sin(2\pi t/T_{\rm w})$, with a sample interval $\Delta= 1\;{\rm ms}$, and $T_{\rm w}= 1.024\;{\rm s}$, so that exactly 100 periods fit into the sampling interval. Plot the magnitude of the DFT of this dataset as a function of $\omega$ using a logarithmic scale for the vertical axis, and comment briefly on it.

```python
import numpy
```

Define the sampling interval, $\Delta$, and the number of samples, $N$.

```python
Delta, N = 1e-3, 1024
```

With the number of samples and the sampling interval, we can find the sampling time:

```python
T = N*Delta
```

To have 100 periods in our dataset, the waveform period needs to be

```python
Tw = T / 100.0
```

Create list of sampling times and sample data.

```python
t_list = numpy.linspace( 0, T, N )
y_list = numpy.sin( 2*numpy.pi*t_list/Tw )
```

```python
import matplotlib.pyplot as pyplot
```

```python
# tell ipython to show figure in the notebook
%matplotlib inline
pyplot.plot( t_list, y_list );
pyplot.xlabel( r"$t$", fontsize=16 );
pyplot.axis( [t_list[0], t_list[-1], -1, +1] );
```

```python
C_list = numpy.array( [numpy.dot(y_list,numpy.exp(2j*numpy.pi*n*t_list/T)) for n in range(N)] )
```

```python
pyplot.plot( numpy.abs(C_list) )
pyplot.xlim( 0, N )
pyplot.yscale( 'log' )
```

```python
pyplot.plot( numpy.abs(C_list[:N/2]) )
pyplot.xlim( 0, N/2 )
pyplot.yscale( 'log' )
```

To obtain 100.5 cycles, the waveform period needs to be changed. It was $T_{\rm w}= T/100$ but now it is:

```python
Tw_g = T / 100.5
```

So calculate the new dataset and plot the last couple cycles to see the effect. There is the original dataset that ends at a complete period (blue) and the new dataset that ends with a half cycle (green).
The graph is rather coarse because we only have $N$ sample points and are plotting with $0.02N$ sample points.

```python
yg_list = numpy.sin( 2*numpy.pi*t_list/Tw_g )
pyplot.plot( t_list, y_list, 'b', t_list, yg_list, 'g' )
pyplot.xlim( 0.98*T, T );
```

Compute the DFT and plot the result.

```python
Cg_list = numpy.array( [numpy.dot(yg_list,numpy.exp(2j*numpy.pi*n*t_list/T)) for n in range(N)] )
pyplot.plot( numpy.abs(Cg_list[:N/2]) )
pyplot.xlim( 0, N/2 )
pyplot.yscale( 'log' )
```

Considering only the positive frequencies, we still have only a single peak. But the contributions from other values of $n$ are significant. They were on the order of $10^{-12}$ whereas now the least is greater than $10^{-1}$.

```python
hanning = 0.5*( 1.0 - numpy.cos( 2*numpy.pi*t_list/T ) )
ynew_list = hanning * yg_list
Cnew_list = numpy.array( [numpy.dot(ynew_list,numpy.exp(2j*numpy.pi*n*t_list/T)) for n in range(N)] )
pyplot.plot( numpy.abs(Cnew_list[:N/2]) )
pyplot.xlim( 0, N/2 )
pyplot.yscale( 'log' )
```

So the other contributions have been significantly reduced. However, we have broadened our desired peak. To see this, let's plot all three sets of Fourier coefficients, clipping the plot range.

```python
pyplot.plot( numpy.abs(C_list[:N/2]), label='original' );
pyplot.plot( numpy.abs(Cg_list[:N/2]), label='glitch' );
pyplot.plot( numpy.abs(Cnew_list[:N/2]), label='hanning' );
pyplot.xlim( 0, N/4 );
pyplot.ylim( 1, 1e3 );
pyplot.yscale( 'log' );
pyplot.legend( fontsize=14 );
```

In this form, we can see more clearly that the "glitch" added significant contributions away from the actual signal frequency. Windowing the dataset corrected this, but broadened the peak.
